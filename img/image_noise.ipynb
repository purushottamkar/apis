```python
import numpy as np
from imglib import noiseSP, rescale
import cv2
from skimage.metrics import peak_signal_noise_ratio
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import os
```

```python
frac_k=0.2

input_dir='input/Set12'
output_dir='input/salt_pepper_noise/'

for im in os.listdir(input_dir):
    img=cv2.imread(os.path.join(input_dir,im),0)
    #add  salt-pepper noise
    noise_type=3
    noisy_img=noiseSP(img.copy(), frac_k, noise_type)
    noisy_psnr=peak_signal_noise_ratio( img, noisy_img)
    noisy_ssim = ssim(img, noisy_img )
    cv2.imwrite(output_dir+im,noisy_img)
    
    plt.imshow(noisy_img, 'gray')
    plt.show()
    
    print(f'psnr={noisy_psnr:.2f},      ssim={noisy_ssim:.2f}')
```

```python
#Block Noise

for i in range(1,13): 
    img1=cv2.imread('input/Set12/'+str(i)+'.png',0)
    img2=cv2.imread('input/checker/checker_512_4.png',0)
    
    
    assert img1.shape == img2.shape
    noisy_img=rescale(0.5*img1+0.5*img2)
    
    noisy_psnr=peak_signal_noise_ratio( img1, noisy_img)
    noisy_ssim = ssim(img1, noisy_img )
    
    cv2.imwrite('input/block_noise/'+str(i)+'.png',noisy_img)
    
    plt.imshow(noisy_img, 'gray')
    plt.show()
    
    print(f'psnr={noisy_psnr:.2f},      ssim={noisy_ssim:.2f}')
```
